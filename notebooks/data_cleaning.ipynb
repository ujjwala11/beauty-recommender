```python
review_category_cols = [
    "skin_tone",
    "eye_color",
    "skin_type",
    "hair_color"
]
```

```python
product_category_cols = [
    "primary_category",
    "secondary_category",
    "tertiary_category"
]
```

```python
def normalize_text_columns(df, columns):
    """
    Normalize categorical text columns by:
    - removing leading/trailing spaces
    - collapsing multiple spaces
    """

    for col in columns:
        df[col] = (
            df[col]
            .str.strip()
            .str.replace(r"\s+", " ", regex=True)
        )

    return df
```

Missing Value Treatment:
three kinds:
1. Truly Missing (Data Collection Issue)
2. Not Applicable (Business Meaning)
3. Unknown

| Column            | Why Missing?                         | Action                     | Why?                                    |
| ----------------- | ------------------------------------ | -------------------------- | --------------------------------------- |
| sale_price_usd    | Product not on sale                  | Keep NaN                   | NaN has business meaning                |
| value_price_usd   | No value pack                        | Keep NaN                   | Business meaning                        |
| variation_desc    | Product has no variation description | Fill with `"Not Provided"` | Easier downstream text processing       |
| size              | Some products don't have a size      | Keep NaN                   | Not applicable                          |
| variation_type    | Some products have no variants       | Keep NaN                   | Business meaning                        |
| variation_value   | Same                                 | Keep NaN                   | Business meaning                        |
| ingredients       | Missing source data                  | Keep NaN                   | We'll handle in Feature Engineering     |
| highlights        | Missing marketing text               | Fill with `"Not Provided"` | Safe for NLP                            |
| tertiary_category | Product only has two levels          | Keep NaN                   | Business meaning                        |
| child_max_price   | No child products                    | Keep NaN                   | Business meaning                        |
| child_min_price   | No child products                    | Keep NaN                   | Business meaning                        |
| rating            | No ratings yet                       | Keep NaN                   | New products shouldn't get fake ratings |
| reviews           | No reviews yet                       | Keep NaN                   | Preserve meaning                        |

```python
def clean_products(df):
    """
    Clean the products dataset using business-aware rules.
    """

    df = df.copy()

    # Text columns where missing means "not provided"
    text_empty = [
        "variation_desc",
        "highlights",
        "ingredients"
    ]

    for col in text_empty:
        df[col] = df[col].fillna("")

    # Optional categorical fields
    unknown_cols = [
        "size",
        "variation_type",
        "variation_value",
        "secondary_category",
        "tertiary_category"
    ]

    for col in unknown_cols:
        df[col] = df[col].fillna("Unknown")

    # Leave business metrics unchanged
    # rating
    # reviews
    # sale_price_usd
    # value_price_usd
    # child_min_price
    # child_max_price

    return df
```

```python
products_clean = clean_products(products_df)

```

```python
products_clean.isna().sum()
```

```python
def clean_reviews(df):
    """
    Clean review dataset using business-aware rules.
    """

    df = df.copy()

    # Optional text
    df["review_text"] = df["review_text"].fillna("")
    df["review_title"] = df["review_title"].fillna("")

    # Optional user profile information
    unknown_cols = [
        "skin_type",
        "skin_tone",
        "eye_color",
        "hair_color"
    ]

    for col in unknown_cols:
        df[col] = df[col].fillna("Unknown")

    # Keep these unchanged
    # helpfulness
    # is_recommended

    return df
```

```python
reviews_clean = clean_reviews(review_dfs)



```

```python
def clean_ingredients(df):
    """
    Clean ingredient knowledge base.
    """

    df = df.copy()

    df["avoid_with"] = df["avoid_with"].fillna("")

    return df
```

In real-world recommendation systems, multiple reviews from the same user on the same product can represent evolving opinions. However, for modeling user preference, we want a single stable signal per interaction pair. Keeping the most recent review preserves updated sentiment while avoiding inflated user-product weighting and bias in collaborative filtering models.
